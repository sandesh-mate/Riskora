```python
import yfinance as yf
import pandas as pd
import os

# Folder paths (matches your project folder)
raw_data_path = "02_Data/Raw_Data/"
backup_data_path = "02_Data/Backup_Data/"

# Create folders if not exist
os.makedirs(raw_data_path, exist_ok=True)
os.makedirs(backup_data_path, exist_ok=True)

# Stock list (from your project)
stocks = {
    "TCS": "TCS.NS",
    "Infosys": "INFY.NS",
    "HDFC_Bank": "HDFCBANK.NS",
    "ICICI_Bank": "ICICIBANK.NS",
    "Maruti": "MARUTI.NS",
    "Reliance": "RELIANCE.NS",
    "Apple": "AAPL",
    "Amazon": "AMZN"
}

# Date range (project requirement)
start_date = "2023-01-01"
end_date = "2025-12-31"

# Download data with error handling
for company, ticker in stocks.items():
    try:
        print(f"Downloading {company} data...")
        
        data = yf.download(ticker, start=start_date, end=end_date)
        
        # Keep required columns safely (fix for Adj Close issue)
        columns_needed = ['Open', 'Close', 'Volume', 'Adj Close']
        available_columns = [col for col in columns_needed if col in data.columns]
        data = data[available_columns]
        
        # Save raw file
        raw_file = raw_data_path + company + "_raw.csv"
        data.to_csv(raw_file)
        
        # Save backup file
        backup_file = backup_data_path + company + "_backup.csv"
        data.to_csv(backup_file)
        
        print(f"✅ {company} saved successfully")
        
    except Exception as e:
        print(f"❌ Error downloading {company}: {e}")

print("🎉 All stock data downloaded and saved!")
```
